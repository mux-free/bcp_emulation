```python
for month in month_list:

    print(f'\nStart with {month}')

    MONTHS = ['JAN','FEB','MAR','APR','MAY','JUN','JUL','AUG','SEP','OCT','NOV','DEC']
    MONTHSN = np.arange(1,13,1)

    sp = '/net/helium/atmosdyn/freimax/data_msc/IFS-18/IFS-traj/CYC_validation'
    f = open(f'{sp}/{month}/Validation-CYC-randfo-{month}.txt','rb')
    gd = pickle.load(f)
    f.close()

    Lat = np.round(np.linspace(0,90,226),2)
    Lon = np.round(np.linspace(-180,180,901),2)

    for k in gd.keys():
        if k == month:
            for j in gd[k].keys():
                mid = np.where(gd[k][j]['hSLP']==0)[0][0]
                clat = gd[k][j]['clat'][mid].astype(int)
                clon = gd[k][j]['clon'][mid].astype(int)
                t = gd[k][j]['dates'][mid]

                yyyy = int(t[0:4])
                MM = int(t[4:6])
                DD = int(t[6:8])
                hh = int(t[9:])

                print(t)
                print(f'/net/helium/atmosdyn/IFS-1Y/{month}/cdf/S{t}')
                s = xr.open_dataset(f'/net/helium/atmosdyn/IFS-1Y/{month}/cdf/S{t}', drop_variables=['PV','P','TH','THE','RH','PVRCONVT','PVRCONVM','PVRTURBT','PVRTURBM','PVRLS','PVRCOND','PVRSW','PVRLWH','PVRLWC','PVRDEP','PVREVC','PVREVR','PVRSUBI','PVRSUBS','PVRMELTI','PVRMELTS','PVRFRZ','PVRRIME','PVRBF'])
                # s = xr.open_dataset('/net/thermo/atmosdyn2/atroman/PACaccu/cdf/P'+t, drop_variables=['ttot','tdyn','tsw','tlw','tmix','tconv','tls','tcond','tdep','tbf','tevc','tfrz','trime','tce'])

                pt = np.array([])
                plat = np.array([])
                plon = np.array([])
                PS = s.PS.values[0,0,clat,clon]

                for l in range(len(clat)):
                    P = helper.modellevel_to_pressure(PS[l])
                    pid = np.where((P>=700) & (P<=975))[0] #& (pv[:,clat[l],clon[l]]>=0.75))
                    for i in pid:
                            pt = np.append(pt,P[i])
                            plat = np.append(plat,Lat[clat[l]])
                            plon = np.append(plon,Lon[clon[l]])
                            # print(clat[l])

                save = np.zeros((len(pt),4))
                save[:,1] = plon
                save[:,2] = plat
                save[:,3] = pt

                print('Done, Now save text')
                # Create the directory if it does not exist
                
                PATH = f'{sp}/{month}/ctraj/'
                os.makedirs(PATH, exist_ok=True)
                
                np.savetxt(PATH + 'trastart-mature-' + t + '-ID-' + '%06d'%j + '.txt',save,fmt='%f', delimiter=' ', newline='\n')
                print(PATH + 'trastart-mature-' + t + '-ID-' + '%06d'%j + '.txt')
```
